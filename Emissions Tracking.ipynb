# Modeling and Evaluation

```python
# Import necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression

from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import GradientBoostingRegressor

from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import numpy as np
import warnings

# To disable warning
warnings.filterwarnings('ignore')
```

```python
# Checking dataset shape and target variable
df['Per Capita'].nunique()
df.shape

# Selecting Target
y = df['Per Capita']

# Choosing Training and Testing dataset
X = df.drop(['Per Capita','ISO 3166-1 alpha-3','Country'], axis=1)

# Split into training and testing with split ratio of 20%
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state=0)
```

```python
# Scaling the features for all models
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

   ## Linear Regression

```python
# Linear Regression
lr = LinearRegression()
lr.fit(X_train_scaled, y_train)
lr_pred = lr.predict(X_test_scaled)
lr_mse = mean_squared_error(y_test, lr_pred)
lr_rmse = np.sqrt(lr_mse)
print("Linear Regression MSE:", lr_mse)
print("Linear Regression RMSE:", lr_rmse)

# Plot Actual vs Predicted values for Linear Regression
plt.figure(figsize=(16,10))
plt.scatter(y_test, lr_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--') # y=x line
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Linear Regression Predictions')
plt.show()
```

## Decision Tree

```python
# Decision Tree Regressor
dt = DecisionTreeRegressor(random_state=0)
dt.fit(X_train_scaled, y_train)
dt_pred = dt.predict(X_test_scaled)
dt_mse = mean_squared_error(y_test, dt_pred)
dt_rmse = np.sqrt(dt_mse)
print("Decision Tree Regressor MSE:", dt_mse)
print("Decision Tree Regressor RMSE:", dt_rmse)

# Plot Actual vs Predicted values for Decision Tree Regressor
plt.figure(figsize=(16,10))
plt.scatter(y_test, dt_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--') # y=x line
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Decision Tree Regressor Predictions')
plt.show()
```

## Random Forest

```python
# Random Forest Regressor
rf = RandomForestRegressor(random_state=0)
rf.fit(X_train_scaled, y_train)
rf_pred = rf.predict(X_test_scaled)
rf_mse = mean_squared_error(y_test, rf_pred)
rf_rmse = np.sqrt(rf_mse)
print("Random Forest Regressor MSE:", rf_mse)
print("Random Forest Regressor RMSE:", rf_rmse)

# Plot Actual vs Predicted values for Random Forest Regressor
plt.figure(figsize=(16,10))
plt.scatter(y_test, rf_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--') # y=x line
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Random Forest Regressor Predictions')
plt.show()
```

## SVR

```python
# Support Vector Regression (SVR)
svr = SVR()
svr.fit(X_train_scaled, y_train)
svr_pred = svr.predict(X_test_scaled)
svr_mse = mean_squared_error(y_test, svr_pred)
svr_rmse = np.sqrt(svr_mse)
print("Support Vector Regression MSE:", svr_mse)
print("Support Vector Regression RMSE:", svr_rmse)

# Plot Actual vs Predicted values for SVR
plt.figure(figsize=(16,10))
plt.scatter(y_test, svr_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--') # y=x line
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Support Vector Regression Predictions')
plt.show()
```

## Gradient Boosting Regressor

```python
# Gradient Boosting Regressor
gb = GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, max_depth=3, random_state=0)
gb.fit(X_train_scaled, y_train)
gb_pred = gb.predict(X_test_scaled)
gb_mse = mean_squared_error(y_test, gb_pred)
gb_rmse = np.sqrt(gb_mse)
print("Gradient Boosting MSE:", gb_mse)
print("Gradient Boosting RMSE:", gb_rmse)

# Plot Actual vs Predicted values for Gradient Boosting
plt.figure(figsize=(16,10))
plt.scatter(y_test, gb_pred)
plt.plot([min(y_test), max(y_test)], [min(y_test), max(y_test)], color='red', linestyle='--') # y=x line
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Gradient Boosting Predictions')
plt.show()
```
